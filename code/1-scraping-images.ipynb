```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import os
import urllib
import urllib.request
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from google_images_download import google_images_download
```

```python
#open page
chromedriver = "C:/Users/yichi/Desktop/Metis/Lectures/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver

# driver = webdriver.Chrome(chromedriver)


# function to scrape all product images on first page
master_url = 'https://www.wayfair.com/furniture/sb0/'

def scrape_first_product_page(rest_url, local_filename):
    url = master_url + rest_url
    driver = webdriver.Chrome(chromedriver)
    driver.get(url)

    time.sleep(3)

    soup = BeautifulSoup(driver.page_source, "lxml")
    xpath = soup.find_all('img', class_='ImageComponent-image ImageComponent-image--preventVerticalPopping')
    image_urls = [component['src'] for component in xpath]

    # download images from links
    image_save_path = 'scraped_furniture_images/' + local_filename
    for url in image_urls:
        urllib.request.urlretrieve(url, image_save_path + url.split('/')[-1])


# function to scrape product images on rest pages
def scrape_rest_product_pages(rest_url, scrape_ranges, local_filename):
#     all_pages = range(54,61)  # pattern starts from second page. Each page has 40-50 product labels.

    for page in scrape_ranges:
        url = master_url + rest_url + '.html?curpage={}'.format(page)
#       url = 'https://www.wayfair.com/furniture/sb0/kitchen-dining-chairs-c46130.html?curpage={}'.format(page)
        driver = webdriver.Chrome(chromedriver)
        driver.get(url)

    #   let page get fully loaded
        time.sleep(25)

        soup = BeautifulSoup(driver.page_source, "lxml")

        time.sleep(4)

        xpath = soup.find_all('img', class_='ImageComponent-image ImageComponent-image--preventVerticalPopping')
        image_urls = [component['src'] for component in xpath]

        # download images from links
        image_save_path = 'scraped_furniture_images/' + local_filename
        for url in image_urls:
            urllib.request.urlretrieve(url, image_save_path + url.split('/')[-1])
```
