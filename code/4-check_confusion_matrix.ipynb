```python
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.metrics import accuracy_score,precision_score, recall_score
from sklearn.metrics import precision_recall_curve, confusion_matrix,f1_score, fbeta_score, roc_auc_score
```

```python
path = 'C:/Users/yichi/Desktop/Metis/Project5-Image-Processing/'
with open(path + "pickle/labels.pkl", 'rb') as picklefile: 
    labels = pickle.load(picklefile)
with open(path + "pickle/pred_list.pkl", 'rb') as picklefile: 
    pred_list = pickle.load(picklefile)
```

```python
# plot confusion matrix
cm = confusion_matrix(labels[:12000], pred_list[:12000])
plt.figure(dpi=150)
sns.heatmap(cm, cmap=plt.cm.Oranges, annot=True, fmt='g',square=True,
           xticklabels=['bean_bag_chairs', 'coffee_tables', 'console_table', \
                         'dining_chairs', 'futons', 'lounge_sofa_chairs', 'recliners', \
                         'rocking_chairs', 'sectionals', 'sofas', 'sofa_slipcovers'],
           yticklabels=['bean_bag_chairs', 'coffee_tables', 'console_table', 
                         'dining_chairs', 'futons', 'lounge_sofa_chairs', 'recliners', 
                         'rocking_chairs', 'sectionals', 'sofas', 'sofa_slipcovers'])

plt.xlabel('Predicted Clicks')
plt.ylabel('Actual Clicks')
plt.xticks(rotation=70)
plt.title('Image Classification Neural Net Confusion Matrix');

plt.savefig("confusion_matrix_nn")
```

```python
f1_score(labels[:12000], pred_list[:12000], average='macro')
```

```python
accuracy_score(labels, pred_list)
```
